```python
import pandas as pd
from prompts import *
import plotly.express as px
import glob

pd.set_option('display.max_colwidth', None)
```

```python
all_dfs = []
for file in ['british_dialects_midlands_female.csv',
 'british_dialects_welsh_male.csv',
 'british_dialects_scottish_female.csv',
 'british_dialects_irish_male.csv']:# glob.glob('british_dialects_*'):
    df = pd.read_csv(file)
    df['area'] = file.split('_')[2]
    df['sex'] = file.split('_')[3][:-4]
    df['content'] = file[17:-4]
    all_dfs.append(df)

full_df = pd.concat(all_dfs)
```

```python
df_adjectives = pd.read_csv('filtered_valence_adjectives.csv')
```

```python
def parse_adjectives(row, col, adjectivelist):
    adj_found = []
    for adj in list(set(adjectivelist)):
        if adj in row[col].lower():
            adj_found.append(adj)

    return adj_found
```
