```python
df_accents = pd.read_csv('english_dataset.csv')
```

```python
df_accents.head()
```

```python
import plotly.express as px

grouped = df_accents.groupby('native_language').count().reset_index()

fig = px.bar(grouped, x='native_language', y='speakerid')
fig.show()
```

```python
natives = df_accents[df_accents['native_language'] == 'english']

grouped = natives.groupby('country').count().reset_index()

fig = px.bar(grouped, x='country', y='speakerid')
fig.show()
```

```python
usa_natives = df_accents[(df_accents['native_language'] == 'english') & (df_accents['country'] == 'usa')]

grouped = usa_natives.groupby('state').count().reset_index()

fig = px.bar(grouped, x='state', y='speakerid')
fig.show()
```

## Look at professions

```python
df_british['parsed_professions'] = df_british['extracted_professions_model_response_profession_Qwen2-Audio-7B-Instruct'].str.replace('assistant\n\n', '')

grouped = df_british.groupby(['speaker_origin', 'parsed_professions']).size().reset_index(name="count")

grouped['percentage'] = (grouped['count'] / grouped.groupby("speaker_origin")["count"].transform("sum") ) *100

grouped = grouped[grouped['percentage'] > 1]

fig = px.bar(
    grouped,
    x="speaker_origin", y="percentage", color="parsed_professions",
    barmode="group",
    title="Results Qwen British Dialects"
)

fig.update_layout(width=2500)

fig.show()
```
